```python
#2. what is the average RAM in all the mobiles
data["ram(GB)"].mean()
```

```python
#3. which phone name is having the max battery
data.loc[data["battery"]==data["battery"].max(),["phone_name","battery"]]
```

```python
#4. which phone name is having the min battery
data.loc[data["battery"]==data["battery"].min(),["phone_name","battery"]]
```

```python
#5. what is the max battery given by Huawei
data.loc[((data["brand"]=="Huawei") & (data["battery"]==data["battery"])),["battery"]].max()
```

```python
#6. which ph brand has the min RAM
data.loc[data["ram(GB)"]==data["ram(GB)"].min(),["brand","ram(GB)"]]
```

```python
#7. which is the oldest phone anounced also with dates of anouncements
data.loc[data["announcement_date"]==data["announcement_date"].min(),["phone_name","announcement_date"]]
```

```python
#8. how many phones has android 7.0 as their operating system
data.loc[data["os"]=="Android 7.0",["phone_name"]].count()
```

```python
#9. which brand is having least no of phones produced 
data.loc[data["phone_name"]==data["phone_name"].count(),["phone_name"]]
```

```python
#10 Give me phone name and brand name having the costliest phone
data.loc[data["price(USD)"]==data["price(USD)"].max(),["phone_name","brand","price(USD)"]]
```
